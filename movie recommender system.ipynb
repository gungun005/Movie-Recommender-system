```python
#this data is list of dictionaries
# we have to convert this into['action','adventure','fantasy','scifi']
#this is the helper function 
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
def convert(obj):
    L=[]
    cnt=0
    for i in ast.literal_eval(obj):
        if(cnt!=3):
            L.append(i['name'])
            cnt+=1
        else:
            break
    return L
```

```python
#jo steming krni thi woh data pe hi krdenge aur library use krenge
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#ab ek helper function bnega..
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#it is recommended to convert the data into lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
# ab aage hme yeh sochno hae ki hm kaise recommend krenge -> on the basis of similiarity 
# data jitna bhi hmare pass hae woh textual data hae (jo kuch bhi hae woh saare text form hae)
#kisi bhi mathematical tarreke sae score nikal paye similiarity ke liye
#vectorization( the text will be converted into vectors)
# har movie vector bn jayega
#closest vectors ko hm pakedenge aur return karayenge
#TECHNIQUES TO CONVERT THE WORDS INTO VECTORS
#->Bag of words
#->wordvec
#->ifidf
# we are using bag of words for our project..
# sare tags ko concatinate krdiya jo ki manlo large text bn jayega aur usme sae hm llet krte hae ki hme 5000
#most frequent /commonly used words chahiye 
#har word k frequency count krenge
#jinka top 5000 topmost ko extract krenge by frequency
# fir hm check krenge har movie ki tags mae jo ki words hmne extract kiye the kauns awork kitni baar aya hae
#in vectorization we will not use stop words
#stop words which are used for sentence formation unka khud ka koi meaning nhi hota
#such and,an ,in,or,are etc
#hm use karenge SCIKIT library jisme class hota hae count vectorizer
from  sklearn.feature_extraction.text import CountVectorizer
#isme bht features hote hae lekin hm max features lenge and stop words
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#ab hm apne object cv ko use krenge aur ise explicitly convert krenge into numpy array
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
#the best part of this class is that we can see what is 5000 most frequent words
cv.get_feature_names_out()
```

```python
# ab un words ko htana hae jaise accepted,accepts,accept ke hi saare alag alag hae mtlb
#jiske liye hm steming technique apply karenge..
#['loves','loved','love']->['love','love','love'] hojayega
ps.stem('loved')
```

```python
#4806 movies ke 4806 vectors hae jisme 5000 elements hae..har movie ek vector har movie k bki ke saath distn calculayte distance xyaada similiarity kaam
#we will calculate the cosine distance angle nikalenge..
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#ab ek function bnana hae recommend bolke toh agar koi movie deta hae toh yeh functn five similiar movies btaye
#jaise avtar movie hae uska index nikalenge 
#fir us index ke respect mae similarity array nikalnege
#fir use sort krenge
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

recommend('Avatar')

```python
recommend('Batman Begins')
```
